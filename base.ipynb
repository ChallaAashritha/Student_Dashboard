```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
#reading the data
data = pd.read_excel('processed_student_data.xlsx', sheet_name='in')
data.head(5)
```

```python
#target feature
target=data['G3']
target
```

```python
#input features
features=data[['traveltime','studytime','failures','schoolsup','famsup','paid','activities','internet','freetime','goout','Dalc','Walc','health','absences','G1','G2','sub','sex']]
```

```python
#viewing the features data
features.head(5)
```

```python
#converting categorical to numerical one
le = LabelEncoder()
for col in ['schoolsup','famsup','paid','activities','internet','sub']:
    features[col] = le.fit_transform(features[col])
```

```python
features.head(5)
```

```python
target.head(5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.2, random_state=42)
linearreg=LinearRegression()
```

```python
linearreg.fit(X_train,y_train)
```

```python
y_pred=linearreg.predict(X_test)
```

```python
print("R² Score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
# Predictions for train and test separately
y_train_pred = linearreg.predict(X_train)   
y_test_pred  = linearreg.predict(X_test)  
# Training performance
print("Train R²:", r2_score(y_train, y_train_pred))
print("Train RMSE:", np.sqrt(mean_squared_error(y_train, y_train_pred)))
# Testing performance
print("Test R²:", r2_score(y_test, y_test_pred))
print("Test RMSE:", np.sqrt(mean_squared_error(y_test, y_test_pred)))
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
import numpy as np
import pandas as pd

# Binary target
y_bin = (target >= 7.2).astype(int)

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
logreg = LogisticRegression(max_iter=1000, class_weight="balanced")  # handle class imbalance

# Lists to store metrics
accs_train, accs_test = [], []
precisions, recalls, f1s, aucs = [], [], [], []
conf_matrices = []

for fold, (train_idx, test_idx) in enumerate(skf.split(features, y_bin), 1):
    X_train, X_test = features.iloc[train_idx], features.iloc[test_idx]
    y_train, y_test = y_bin.iloc[train_idx], y_bin.iloc[test_idx]
    
    logreg.fit(X_train, y_train)
    y_pred_test = logreg.predict(X_test)
    y_pred_train = logreg.predict(X_train)
    y_proba_test = logreg.predict_proba(X_test)[:,1]
    
    # Train and Test Accuracy
    accs_train.append(accuracy_score(y_train, y_pred_train))
    accs_test.append(accuracy_score(y_test, y_pred_test))
    
    # Other metrics
    precisions.append(precision_score(y_test, y_pred_test))
    recalls.append(recall_score(y_test, y_pred_test))
    f1s.append(f1_score(y_test, y_pred_test))
    aucs.append(roc_auc_score(y_test, y_proba_test))
    
    # Confusion matrix
    conf_matrices.append(confusion_matrix(y_test, y_pred_test))
    print(f"\nFold {fold} Confusion Matrix:\n{conf_matrices[-1]}")

# Aggregate results
print("\n===== Logistic Regression Metrics (Stratified KFold 5) =====")
print(f"Train Accuracy: {np.mean(accs_train):.4f}")
print(f"Test Accuracy: {np.mean(accs_test):.4f}")
print(f"Precision: {np.mean(precisions):.4f}")
print(f"Recall: {np.mean(recalls):.4f}")
print(f"F1 Score: {np.mean(f1s):.4f}")
print(f"ROC-AUC: {np.mean(aucs):.4f}")
```

```python
import joblib

# Save the Random Forest Regressor model
joblib.dump(linearreg, 'base_lin.pkl')
joblib.dump(logreg, 'base_log.pkl')
```
